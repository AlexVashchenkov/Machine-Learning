```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import os
import requests
```

```python
output_dir = "cat_images"

def parse_category(driver, category_name, category_length):
    loaded_images = 0
    
    breed_dir = os.path.join(output_dir, category_name)
    os.makedirs(breed_dir, exist_ok=True)

    scroll_div = driver.find_element(By.XPATH, "/html/body/main/div[1]/div/div[5]/div[2]/div/div[2]/div/div[5]/div[2]/div/div/div[1]/div/div[3]")
    
    for pages in range(1, (category_length - 1758) // 30 + 1):
        
        for i in range(1, 31):
            image = driver.find_element(By.XPATH, f"/html/body/main/div[1]/div/div[5]/div[2]/div/div[2]/div/div[5]/div[2]/div/div/div[1]/div/div[3]/div[1]/div[{loaded_images + i}]/img")
            driver.execute_script("arguments[0].scrollIntoView(true);", image)
            img_data = requests.get(image.get_attribute("src")).content
            file_path = os.path.join(breed_dir, f"{loaded_images + i}.jpg")
            with open(file_path, "wb") as f:
                f.write(img_data)
            print(f"Сохранено: {file_path}")
            
        loaded_images += 30
        load_more = driver.find_element(By.CLASS_NAME, "sc-drHDoh.fQpLKW")
        driver.execute_script("arguments[0].click();", load_more)
        driver.execute_script("arguments[0].scrollTo(0, arguments[0].scrollHeight);", scroll_div)
```

```python
category_numbers = [49]

for number in category_numbers:
    category = driver.find_element(By.XPATH, f"/html/body/main/div[1]/div/div[5]/div[2]/div/div[2]/div/div[5]/div[2]/div/div/div[2]/div[1]/div/ul/li[2]/ul/li[{number}]/div/div[3]/div/p")
    category_name = driver.find_element(By.XPATH, "/html/body/main/div[1]/div/div[5]/div[2]/div/div[2]/div/div[5]/div[2]/div/div/div[1]/div/div[1]/h2")
    driver.execute_script("arguments[0].scrollIntoView(true);", category_name)
    category.click()
    category_name_div = driver.find_element(By.XPATH, "/html/body/main/div[1]/div/div[5]/div[2]/div/div[2]/div/div[5]/div[2]/div/div/div[1]/div/div[1]/h2")
    number_of_items_div = driver.find_element(By.CLASS_NAME, "sc-ovuCP.sc-kwXLfY.jgWmzR.gTyLYZ").text

    category_name = category_name_div.text.split('(')[0]
    number_of_items = int(number_of_items_div.split(' ')[0][1:])
    if number_of_items > 1000:
        print(category_name, number_of_items)
        parse_category(driver, category_name, number_of_items)
```
